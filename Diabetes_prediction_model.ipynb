```python
data = pd.read_csv('diabetes.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
m1 = data["SkinThickness"].median()
m1
```

```python
for x in data.index:
   if data.loc[x,"SkinThickness"] == 0:
      data.loc[x,"SkinThickness"] = m1
data
```

```python
data.iloc[51:54]
```

```python
data.loc[(data["Insulin"] >= 20) & (data["Insulin"] <= 50), "Insulin"] = 50
data
```

```python
data.iloc[51:54]
```

```python
data.loc[(data["Insulin"] > 400 ),"Insulin"] = 400
```

```python
m2=data["Insulin"].median()
m2
```

```python
m3=data["Insulin"].mean()
m3
```

```python
for x in data.index:
   if data.loc[x,"Insulin"]==0:
     data.loc[x,"Insulin"]=m2
data
```

```python
X = data.values[:,0:8]
X
```

```python
Y = data.values[:,-1]
Y
```

```python
#Splitting the data into training and testing data
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
#Fitting Decision Tree classifier to training set
classifier = DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier.fit(X_train,Y_train)
```

```python
#Predicting test values with model
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
cm = confusion_matrix(Y_test,y_pred)
print(cm)
```

```python
#Calculating accuracy
accuracy_score=accuracy_score(Y_test,y_pred)
print(accuracy_score)
```

```python
print(classification_report(Y_test,y_pred))
```
